```python
import numpy as np
import pymc as pm
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import calibration_curve
from sklearn.metrics import accuracy_score
import arviz as az
```

```python
# 1. Create a synthetic binary classification dataset.
X, y = make_classification(n_samples=500, n_features=5, n_informative=3,
                           n_redundant=0, n_classes=2, random_state=42)

# 2. Split data: first 400 samples for training, remaining 100 for testing.
X_train, X_test = X[:400], X[400:]
y_train, y_test = y[:400], y[400:]

# 3. Standardize features (very important for both inference and classical training).
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
with pm.Model() as bayes_model:
    # Priors for intercept (bias) and coefficients (weights)
    intercept = pm.Normal("intercept", mu=0, sigma=1)
    coefs = pm.Normal("coefs", mu=0, sigma=1, shape=X_train_scaled.shape[1])
    
    # Linear predictor: intercept + X * coefficients
    logits = intercept + pm.math.dot(X_train_scaled, coefs)
    
    # Sigmoid transforms logits to probabilities (range (0, 1)).
    p = pm.math.sigmoid(logits)
    
    # Likelihood: observed outcomes modeled via a Bernoulli distribution.
    likelihood = pm.Bernoulli("y", p=p, observed=y_train)
    
    # Sample from the posterior.
    # Using return_inferencedata=False returns a MultiTrace (dictionary-like) object.
    trace = pm.sample(2000, tune=1000, target_accept=0.9, chains=2, cores=1, return_inferencedata=False)
```

```python
# 5. Build an ensemble of 10 logistic regression models from the Bayesian posterior.
#    Extract the posterior samples for intercept and coefficients.
# trace["intercept"] will have shape (n_samples, ) and trace["coefs"] will have shape (n_samples, n_features)
n_samples = trace["intercept"].shape[0]
np.random.seed(42)
ensemble_indices = np.random.choice(n_samples, size=10, replace=False)
ensemble_intercepts = trace["intercept"][ensemble_indices]  # shape (10,)
ensemble_coefs = trace["coefs"][ensemble_indices, :]         # shape (10, n_features)
```

```python
def ensemble_predict(X, intercepts, coefs):
    """Compute predicted probabilities for each ensemble member on data X."""
    ensemble_probs = []
    for i in range(len(intercepts)):
        logits_i = intercepts[i] + np.dot(X, coefs[i])
        probs_i = 1 / (1 + np.exp(-logits_i))
        ensemble_probs.append(probs_i)
    return np.array(ensemble_probs)  # shape: (ensemble_size, n_samples)

# Compute ensemble predictions on the test set.
ensemble_probs = ensemble_predict(X_test_scaled, ensemble_intercepts, ensemble_coefs)
ensemble_pred_probs = ensemble_probs.mean(axis=0)  # Average predictions over the ensemble.
ensemble_pred_class = (ensemble_pred_probs >= 0.5).astype(int)
ensemble_accuracy = accuracy_score(y_test, ensemble_pred_class)
```

```python
def compute_ece(y_true, y_prob, n_bins=10):
    bins = np.linspace(0, 1, n_bins + 1)
    bin_indices = np.digitize(y_prob, bins, right=True)
    ece = 0.0
    for i in range(1, n_bins + 1):
        bin_mask = (bin_indices == i)
        if np.any(bin_mask):
            avg_pred = np.mean(y_prob[bin_mask])
            avg_true = np.mean(y_true[bin_mask])
            bin_weight = np.sum(bin_mask) / len(y_prob)
            ece += np.abs(avg_pred - avg_true) * bin_weight
    return ece

ensemble_ece = compute_ece(y_test, ensemble_pred_probs)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_scaled, y_train)
lr_pred_probs = lr_model.predict_proba(X_test_scaled)[:, 1]
lr_pred_class = lr_model.predict(X_test_scaled)
lr_accuracy = accuracy_score(y_test, lr_pred_class)
lr_ece = compute_ece(y_test, lr_pred_probs)

# Compute calibration curves for both models.
ensemble_frac_pos, ensemble_mean_pred = calibration_curve(y_test, ensemble_pred_probs, n_bins=10)
lr_frac_pos, lr_mean_pred = calibration_curve(y_test, lr_pred_probs, n_bins=10)

# 9. Print performance metrics.
print("Ensemble Accuracy:             {:.3f}".format(ensemble_accuracy))
print("Ensemble ECE:                  {:.3f}".format(ensemble_ece))
print("Logistic Regression Accuracy:  {:.3f}".format(lr_accuracy))
print("Logistic Regression ECE:       {:.3f}".format(lr_ece))

# 10. Plot calibration curves (reliability diagrams) for both models.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(ensemble_mean_pred, ensemble_frac_pos, marker='o', label='Ensemble')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect Calibration')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration Curve - Ensemble')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(lr_mean_pred, lr_frac_pos, marker='o', label='Logistic Regression')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect Calibration')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration Curve - Logistic Regression')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def plot_confidence_curve(y_true, y_pred, folder_path="results", bins=10):

    # Define bin edges based on predictions
    bin_edges = np.linspace(min(y_pred), max(y_pred), bins + 1)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Initialize arrays to store values per bin
    bin_means_pred = []
    bin_means_actual = []
    bin_counts = np.zeros(bins)

    # Assign each prediction to a bin
    bin_indices = np.digitize(y_pred, bin_edges) - 1

    for i in range(bins):
        bin_mask = (bin_indices == i)
        if np.any(bin_mask):
            bin_means_pred.append(np.mean(y_pred[bin_mask]))
            bin_means_actual.append(np.mean(y_true[bin_mask]))
        else:
            bin_means_pred.append(np.nan)
            bin_means_actual.append(np.nan)

    # Convert lists to arrays and handle empty bins
    bin_means_pred = np.array(bin_means_pred)
    bin_means_actual = np.array(bin_means_actual)
    
    # Remove NaN values
    valid_bins = ~np.isnan(bin_means_pred) & ~np.isnan(bin_means_actual)
    bin_means_pred = bin_means_pred[valid_bins]
    bin_means_actual = bin_means_actual[valid_bins]
    bin_centers = bin_centers[valid_bins]

    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(bin_centers, bin_means_actual, marker='o', linestyle='-', color='blue', label="Actual Mean")
    plt.plot(bin_centers, bin_means_pred, marker='s', linestyle='--', color='red', label="Predicted Mean")
    plt.plot(bin_centers, bin_centers, linestyle='dotted', color='black', label="Ideal (y=x)")

    plt.title("Regression Confidence Curve")
    plt.xlabel("Predicted Change (%)")
    plt.ylabel("Actual Change (%)")
    plt.legend()
    plt.grid(True)

    # Ensure folder exists
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Save plot
    plt.savefig(os.path.join(folder_path, "confidence_curve.png"))
    plt.show()
plot_confidence_curve(y_test, bayes_preds_mean, folder_path="results")
plot_confidence_curve(y_test, lin_reg_preds, folder_path="results")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymc as pm
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.calibration import calibration_curve


X, y = make_classification(n_samples=500, n_features=5, n_informative=3,
                           n_redundant=0, n_classes=3, random_state=42)

X_train, X_test = X[:400], X[400:]
y_train, y_test = y[:400], y[400:]

X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
X_train_scaled = (X_train - X_mean) / X_std
X_test_scaled = (X_test - X_mean) / X_std


with pm.Model() as bayes_model:
    intercept = pm.Normal("intercept", mu=0, sigma=1, shape=3)
    coefs = pm.Normal("coefs", mu=0, sigma=1, shape=(5, 3))
    logits = pm.math.dot(X_train_scaled, coefs) + intercept
    p_unclipped = pm.math.softmax(logits)
    p = pm.Deterministic("p", pm.math.clip(p_unclipped, 1e-6, 1 - 1e-6))
    likelihood = pm.Categorical("likelihood", p=p, observed=y_train)
    start_vals = pm.find_MAP()
    print("MAP starting values:", start_vals)
    trace = pm.sample(1000, tune=1000, target_accept=0.9, start=start_vals, cores=1)
```
